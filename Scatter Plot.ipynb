------------------------------------------------Scatter Plot-------------------------------------------------------

A scatter plot is used to show the relationship between two numerical variables by displaying them as points on a 2D plane.  

 Key Points
- Each point represents a pair of values '(x, y)'.  
- Useful for spotting patterns, trends, clusters, or outliers.  
- The slope of the points can indicate positive or negative correlation.  

Use Cases in Data Science / ML
- Visualizing relationships between features (e.g., height vs weight).  
- Checking if two features are correlated before feeding them into a model.  
- Exploring clusters or class separability in datasets.  

In this example, it generates synthetic data where 'y' has a linear relationship with 'x'. The scatter plot helps visualize that linear trend.


```python
import matplotlib. pyplot as plt 
import numpy as np

x = np. random.randn(40)
y = 2 * x + 1 + 0.1 * np. random.randn(40)

plt.scatter(x, y, label="Data points")
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color="red", label="Trendline")
plt.title("Scatter Plot with Trendline")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```
