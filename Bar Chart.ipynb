------------------------------------------------Bar Chart----------------------------------------------------------

A bar chart is used to represent categorical data with rectangular bars, where the length or height of each bar corresponds to the value of that category.  

Key Points
- The x-axis typically shows categories.  
- The y-axis shows the value or count for each category.  
- Bars can be plotted vertically or horizontally.  
- Useful for comparing different groups or categories at a glance.  

Use Cases in Data Science / ML
- Showing class distribution in a dataset.  
- Comparing sales, revenue, or counts across categories.  
- Visualizing frequency of different labels in classification problems.  

Bar charts make it easy to see which categories are larger, smaller, or equal, helping in quick comparisons and insights.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 20]

plt.bar(categories, values, color='g', label='Values')
plt.title('Bar Chart')
plt.xlabel('Categories') 
plt.ylabel('Values')
plt.legend()
plt.show()
```
