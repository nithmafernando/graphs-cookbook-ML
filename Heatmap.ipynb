---------------------------------------------Heatmap---------------------------------------------------------------

A heatmap is a graphical representation of data where individual values are represented by colors.  
It’s especially useful for visualizing matrices, correlations, or frequency data.  

Key Points
- The color intensity shows the magnitude of values.  
- Can be annotated with exact numbers for clarity (`annot=True`).  
- Works best for 2D data such as matrices or pivot tables.  

Use Cases in Data Science / ML
- Visualizing correlation matrices between features.  
- Representing confusion matrices to evaluate classification models.  
- Showing relationships in pivoted/tabular data (e.g., sales by month and region).  

In this example, generate a random 10×10 matrix and plot it as a heatmap.  
Each cell’s color represents the numeric value, and annotations make it easier to read.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = np. random. rand (10, 10)

sns.heatmap(data, annot=True, cmap="coolwarm")
plt. title( 'Heatmap')
plt. show ()
```
