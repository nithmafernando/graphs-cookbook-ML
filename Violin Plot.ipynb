---------------------------------------------Violin Plot-----------------------------------------------------------

A violin plot combines the features of a box plot and a kernel density plot.  
It not only shows summary statistics (median, quartiles, etc.) but also the probability density of the data at different values.  

Key Points
- The white dot represents the median.  
- The thick bar in the center shows the interquartile range (IQR).  
- The shape (wider = more data, thinner = less data) shows the distribution.  
- Useful when comparing distributions between multiple categories.  

Use Cases in Data Science / ML
- Understanding how features are distributed across categories.  
- Comparing class-wise distributions in classification problems.  
- Spotting whether data is skewed, multimodal, or symmetric.  

In this example, it generates three groups of data with different standard deviations and visualized them using violin plots. The shape of each violin shows how spread out or concentrated the values are.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = [np.random.normal(0, std, 100) for std in range(1, 4)]

sns. violinplot(data=data)
plt. title( 'Violin Plot')
plt. xlabel ('Category')
plt.ylabel ('Values')
plt.show()
```
