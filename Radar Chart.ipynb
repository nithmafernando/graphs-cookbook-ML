--------------------------------------------------Radar Chart------------------------------------------------------

A radar chart (or spider chart) is used to compare multivariate data on a circular axis.  
Each variable has its own axis starting from the center, and values are plotted along these axes.  
The points are connected to form a polygon, making it easy to compare patterns.  

Key Points
- Each axis corresponds to a feature/metric.  
- The distance from the center represents the value of that metric.  
- The shape of the polygon shows the strengths and weaknesses across categories.  
- Can be filled with color for better visualization.  

Use Cases in Data Science / ML
- Comparing model performance metrics (accuracy, precision, recall, F1, etc.).  
- Visualizing multi-feature scores (e.g., survey results, player stats, or KPIs).  
- Spotting which areas are strong/weak across different categories.  

In this example, five categories (Aâ€“E) are plotted, and the polygon shape shows how the values compare across these dimensions.

```python
import matplotlib.pyplot as plt 
import numpy as np

labels = np.array(['A','B', ' C', ' D', ' E'])
data = np.array([4, 5, 3, 4, 2])

angles = np. linspace(0, 2 * np.pi, len(labels), endpoint=False)
data = np. concatenate ((data, [data[0]]))
angles = np. concatenate((angles, [angles[0]]))

plt.polar(angles, data, marker='o')
plt.fill(angles, data, alpha=0.25)
plt.title('Radar Chart')
plt. show()
```
