------------------------------------------------Histogram----------------------------------------------------------

A histogram is used to visualize the distribution of a numerical dataset.  
It groups data into intervals called bins and shows how many values fall into each bin.  

Key Points
- X-axis represents the bins/ranges of values.  
- Y-axis represents the frequency/count of values in each bin.  
- Useful for seeing the shape, spread, and skewness of the data.  

Use Cases in Data Science / ML
- Understanding feature distributions before training a model.  
- Checking if data is normally distributed.  
- Detecting outliers or unusual patterns.  

In this example,it generates 1000 random values from a normal distribution and visualize how they are distributed across 20 bins.

```python
import matplotlib.pyplot as plt 
import numpy as np

data = np.random.randn(1000)

plt. hist(data, bins=20, edgecolor='black')
plt. title( 'Histogram') 
plt. xlabel( 'Values')
plt. ylabel( 'Frequency')
plt. show()
```
