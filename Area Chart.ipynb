---------------------------------------------------Area Chart------------------------------------------------------

An area chart is similar to a line chart, but the area under the line is filled with color.  
It is useful for showing how a quantity changes over time or comparing multiple cumulative trends.  

Key Points
- The x-axis usually represents time or an ordered variable.  
- The y-axis shows the values being measured.  
- Filling the area emphasizes the magnitude of change.  
- Multiple areas can be compared (stacked or overlapping).  

Use Cases in Data Science / ML
- Tracking model performance metrics (accuracy, loss) over time.  
- Visualizing cumulative values (like total users, revenue, or transactions).  
- Comparing growth trends between two or more groups.  

In this example, two sets of values ('y1' and 'y2') are plotted, and the area between them is shaded to show the difference.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 25, 30, 35]
y2 = [5, 10, 20, 25, 30]

plt. fill_between(x, y1, y2, color='skyblue',alpha=0.4)
plt. title('Area Chart') 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
