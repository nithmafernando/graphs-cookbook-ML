-----------------------------------------------Line Chart----------------------------------------------------------

A line graph is used to visualize data points that are connected by straight lines.  
Itâ€™s most commonly used to show trends, patterns, or changes over continuous intervals. (like time, sequence, or ordered data).  

Key Points
- The x-axis usually represents a sequence (e.g., time, steps, or ordered categories).  
- The y-axis represents the measured values.  
- Useful for spotting trends, increases/decreases, and fluctuations in data.  

Use Cases in Data Science / ML
- Tracking model metrics such as training/validation loss or accuracy over epochs.  
- Showing stock market prices or sensor readings over time.  
- Visualizing how a variable changes in relation to another ordered variable.  

In the example below, the line connects (x, y) points to show how the values vary.

```python
import matplotlib.pyplot as plt

x=[1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 3]

plt. plot(x, y)
plt. title('Line Chart') 
plt.xlabel ('X-axis')
plt.ylabel('Y-axis')
plt. show()
```
